```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



data = {
    'Questions': [
        'Use of Generative AI Tools',
        'Familiarity with AI Detection Tools',
        'Perception on AI Compromising Academic Integrity',
        'AI-Assisted Work Considered Original',
        'Willingness to Use AI Detection Tools',
        'Concern About Data Privacy',
        'Impact of AI on Creativity',
        'Impact of AI on Confidence'
    ],
    'Categories': [
        ['Yes', 'No'],
        ['Very Familiar', 'Somewhat Familiar', 'Not Familiar'],
        ['Agree', 'Neutral', 'Disagree'],
        ['Yes', 'No', 'It Depends'],
        ['Yes', 'No', 'Unsure'],
        ['Concerned', 'Not Concerned', 'Unsure'],
        ['Enhances', 'Limits', 'No Effect'],
        ['Enhances', 'No Effect', 'Diminishes']
    ],
    'Values': [
        [80.77, 19.23],
        [15.38, 67.95, 16.67],
        [20.51, 69.23, 10.25],
        [39.74, 41.03, 19.23],
        [65.38, 30.77, 3.85],
        [65.38, 25.64, 8.97],
        [62.82, 24.36, 12.82],
        [58.97, 35.90, 5.13]
    ]
}


def create_bar_plot(categories, values, title):
    plt.figure(figsize=(8,6))
    sns.barplot(x=categories, y=values, palette='Blues_d')
    plt.title(title)
    plt.ylabel('Percentage (%)')
    plt.xlabel('')
    plt.ylim(0, 100)
    for index, value in enumerate(values):
        plt.text(index, value + 1, f'{value}%', ha='center')
    plt.show()


# 1. Use of Generative AI Tools
create_bar_plot(
    data['Categories'][0],
    data['Values'][0],
    'Use of Generative AI Tools'
)

# 2. Concern About Data Privacy
create_bar_plot(
    data['Categories'][5],
    data['Values'][5],
    'Concern About Data Privacy When Using AI Detection Tools'
)

# 3. Impact of AI on Creativity
create_bar_plot(
    data['Categories'][6],
    data['Values'][6],
    'Impact of AI Tools on Creativity'
)

# 4. Perception on AI Compromising Academic Integrity
create_bar_plot(
    ['Agree', 'Neutral', 'Disagree'],
    data['Values'][2],
    'Does AI Use Compromise Academic Integrity?'
)

# 5. Willingness to Use AI Detection Tools
create_bar_plot(
    data['Categories'][4],
    data['Values'][4],
    'Willingness to Use AI Detection Tools'
)
```

```python
import matplotlib.pyplot as plt


labels = ['GPT-4o as Detector on ChatGPT text', 'GPT-4o as Detector on Human-written text']
correctly_classified = [80.8, 66.7]
misclassified = [8.98, 30.78]
unclear = [10.22, 2.52]


fig, ax = plt.subplots(figsize=(8, 4))

# stacked bars
ax.barh(labels, correctly_classified, color='green', label='Correctly classified as AI')
ax.barh(labels, misclassified, left=correctly_classified, color='red', label='Misclassified as human')
ax.barh(labels, unclear, left=[correctly_classified[i] + misclassified[i] for i in range(len(labels))], color='orange', label='Unclear')

# add legend and labels
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_xlabel('Percentage')
ax.set_xlim(0, 100)
ax.set_xticks([0, 25, 50, 75, 100])
plt.title('Detection Results by ChatGPT and GPT-4')


plt.show()
```
